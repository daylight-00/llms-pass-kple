```python
from glob import glob
import os
import pickle
```

```python
ques_path = r"ques_fixed_re"
out_path = os.path.basename(ques_path) + "_detailed"
os.makedirs(out_path, exist_ok=True)

saves_1 = sorted(glob(f'{ques_path}/*1.pkl'))
saves_2 = sorted(glob(f'{ques_path}/*2.pkl'))
saves_3 = sorted(glob(f'{ques_path}/*3.pkl'))
saves_4 = sorted(glob(f'{ques_path}/*4.pkl'))
saves_list = [list(pair) for pair in zip(saves_1, saves_2, saves_3, saves_4)]
```

```python
for path_list in saves_list:
    with open(path_list[0], 'rb') as f:
        exam_1 = pickle.load(f)
    f.close()
    with open(path_list[1], 'rb') as f:
        exam_2 = pickle.load(f)
    f.close()
    with open(path_list[2], 'rb') as f:
        exam_3 = pickle.load(f)
    f.close()
    with open(path_list[3], 'rb') as f:
        exam_4 = pickle.load(f)
    f.close()

    exam_1_1 = exam_1[0:20]
    exam_1_2 = exam_1[20:40]
    exam_1_3 = exam_1[40:60]
    exam_1_4 = exam_1[60:80]
    exam_1_5 = exam_1[80:100]

    exam_2_1 = exam_2[0:18]
    exam_2_2 = exam_2[18:36]
    exam_2_3 = exam_2[36:54]
    exam_2_4 = exam_2[54:72]
    exam_2_5 = exam_2[72:90]

    exam_3_1 = exam_3[0:77]
    exam_3_2 = exam_3[77:104]
    exam_3_3 = exam_3[104:122]
    exam_3_4 = exam_3[122:140]

    exam_4_1 = exam_4[0:20]

    year = os.path.basename(path_list[0])[:4]

    with open(f'{out_path}/{year}_1_1.pkl', 'wb') as f:
        pickle.dump(exam_1_1, f)
    f.close()
    with open(f'{out_path}/{year}_1_2.pkl', 'wb') as f:
        pickle.dump(exam_1_2, f)
    f.close()
    with open(f'{out_path}/{year}_1_3.pkl', 'wb') as f:
        pickle.dump(exam_1_3, f)
    f.close()
    with open(f'{out_path}/{year}_1_4.pkl', 'wb') as f:
        pickle.dump(exam_1_4, f)
    f.close()
    with open(f'{out_path}/{year}_1_5.pkl', 'wb') as f:
        pickle.dump(exam_1_5, f)
    f.close()
    
    with open(f'{out_path}/{year}_2_1.pkl', 'wb') as f:
        pickle.dump(exam_2_1, f)
    f.close()
    with open(f'{out_path}/{year}_2_2.pkl', 'wb') as f:
        pickle.dump(exam_2_2, f)
    f.close()
    with open(f'{out_path}/{year}_2_3.pkl', 'wb') as f:
        pickle.dump(exam_2_3, f)
    f.close()
    with open(f'{out_path}/{year}_2_4.pkl', 'wb') as f:
        pickle.dump(exam_2_4, f)
    f.close()
    with open(f'{out_path}/{year}_2_5.pkl', 'wb') as f:
        pickle.dump(exam_2_5, f)
    f.close()

    with open(f'{out_path}/{year}_3_1.pkl', 'wb') as f:
        pickle.dump(exam_3_1, f)
    f.close()
    with open(f'{out_path}/{year}_3_2.pkl', 'wb') as f:
        pickle.dump(exam_3_2, f)
    f.close()
    with open(f'{out_path}/{year}_3_3.pkl', 'wb') as f:
        pickle.dump(exam_3_3, f)
    f.close()
    with open(f'{out_path}/{year}_3_4.pkl', 'wb') as f:
        pickle.dump(exam_3_4, f)
    f.close()

    with open(f'{out_path}/{year}_4_1.pkl', 'wb') as f:
        pickle.dump(exam_4_1, f)
    f.close()
```
