```python
import pandas as pd

df = pd.read_csv(r"answer_paper.csv", index_col=0)
df
```

```python
column_pairs_by_year = {}

for col in df.columns:
    match = pd.Series(col).str.extract(r'(\d{4})_(\d{2})_(\d{1})')  # Updated regular expression pattern
    if not match.empty:
        year, _, test_no = match.values[0]  # Extract all three values from the match
        if test_no in ['3', '4']:  # 3번 또는 4번 시험지에 해당하는 경우
            if year not in column_pairs_by_year:
                column_pairs_by_year[year] = {'3': [], '4': []}
            column_pairs_by_year[year][test_no].append(col)

# 각 연도에서 3번과 4번 시험지 컬럼이 모두 있는 경우만 쌍으로 추출
column_pairs = []
for year, tests in column_pairs_by_year.items():
    if tests['3'] and tests['4']:  # 3번과 4번 시험지 컬럼이 모두 존재하는 경우
        column_pairs.extend([(col_3, col_4) for col_3 in tests['3'] for col_4 in tests['4']])

# 결과 확인
column_pairs
```

```python
df_expanded = pd.DataFrame(index=range(1, 141))
df_expanded = pd.concat([df_expanded, df], axis=1)
for column_3, column_4 in column_pairs:
    print(column_3, column_4)
    column_3_extended = pd.concat([pd.Series(df[column_3].iloc[:77]), pd.Series(df[column_4]).iloc[:63]], ignore_index=True)
    column_3_extended.index = range(1, 141)

    column_4_adjusted = pd.Series(df[column_4]).iloc[63:83].reset_index(drop=True)
    column_4_adjusted.index = range(1, 21)

    df_expanded[column_3] = column_3_extended
    df_expanded[column_4] = column_4_adjusted
df_expanded.to_csv("answer.csv")
```
